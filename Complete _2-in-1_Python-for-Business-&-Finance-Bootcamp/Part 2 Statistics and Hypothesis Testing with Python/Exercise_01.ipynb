12. Create a __histogram__ with __absolute frequencies__ for the __population__ (75 bins).

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")


import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (12,8), label = "Histogram of Population")
plt.hist(pop , bins = 75)
plt.legend()
plt.show()
```

```python
sns.histplot(pop,
             kde = True,
             color ="blue",
             line_kws = {'color':'blue','linestyle': 'dashed'},
             label = "Histogram")
plt.legend()
sns.despine()
```

```python
sns.histplot(sample ,label = "Histogram Bars")
sns.kdeplot(sample , color ="red", label = "kde line")
```

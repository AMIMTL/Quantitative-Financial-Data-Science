## Exercise 6: How to create great graphs with Matplotlib

1. Before you can start, you have to __import__ matplotlib.pyplot!

```python
import matplotlib.pyplot as plt
```

You consider an investment with the following cashflows:

- -5,000 USD today (Investment)
- 1,200 USD in 1 year
- 1,500 USD in 2 years
- 2,000 USD in 3 years
- 2,500 USD in 4 years
- 3,000 USD in 5 years
- 3,500 USD in 6 years

```python
# run the cell!
cf = [-5000, 1200, 1500, 2000, 2500, 3000, 2500]
cf
```

2. Create one __list with cumulative cashflows__ and one __list with timestamps__!

```python
cf_cum = []
cumul_sum = 0
for i in cf:
  cumul_sum += i
  cf_cum.append(cumul_sum)
print(cf_cum)
```

```python
x_axis = list(range(len(cf)))
x_axis
```

3. Visualize __cumulative cashflows__ in a simple __line plot__! (figure size: (12, 8)) (see result below)

```python
plt.figure(figsize=(12,8))
plt.plot(x_axis , cf_cum)
plt.show()
```

4. Visualize __cumulative cashflows__ in a simple __scatter plot__! (figure size: (12, 8)) (see result below)

```python
plt.figure(figsize =(12,8))
plt.scatter(x_axis ,cf_cum)
plt.show()
```

5. __Customize__ the plot! (see final result below)

```python
#complete the code!
plt.figure(figsize = (12, 8))
plt.plot(x_axis,cf_cum , linestyle ="-.", color ='red', marker ='D', markersize =10, label ="Cumulative Cashflows")
plt.hlines(y = 0 , xmin = x_axis[0], xmax= x_axis[-1], label ="Break Even")
plt.annotate("Break Even Point",
             fontsize = 15,
             xy =(3+0.1,0),
             xytext=(2.5,2000),
             arrowprops={"color":"black"} )
plt.title("Cumulative Cashflow Projrct B", fontsize =15)
plt.xticks(x_axis ,labels = ['t'+str(i) for i in list(range(1,7+1))])
plt.xlabel("Time (in Years)", fontsize = 15)
plt.ylabel("CF in USD", fontsize = 15)
plt.grid()
plt.legend(fontsize =15)
plt.show()
```

6. The ABC Government issued a 30Y Government Bond one year ago with a Coupon Rate of 5.5% (annual payments in arrears). Today´s Bond Price is 97.2 (per 100 par Value). Calculate the Bond´s __current YTM__.

```python
coupon_rate = 5.5/100
par_value = 100
coupon_payment = par_value * coupon_rate
maturity = 30
current_bond_price = 97.2

# calculate ytm
```

```python
cf = [coupon_payment] * maturity
cf[-1] = coupon_payment + par_value
cf.insert(0, -current_bond_price)
print(f"cf: {cf}", sep = " ")
```

```python
targeted_ytm = 5.5/100 # required rate of return %6
step = 0.000001
target_npv_bond = 0 # current_bond_price - total_pv
guess_ytm = targeted_ytm
tolerance = 0.001
```

```python
while True:
    total_pv = 0
    for i in range(1,len(cf)):
        total_pv  += cf[i]/(1+ guess_ytm)**i
    diff = total_pv - current_bond_price
    
    if abs(diff) > tolerance:
        if diff<0:
            guess_ytm -= step
        else:
            guess_ytm += step
    else:
        break

print(f"The total NPV  (Net Present Value) is {total_pv:.4f}." ,  f"The YTM is {guess_ytm *100:.2f}%", sep = '\n')

if guess_ytm < targeted_ytm:
    print("We accept the project !")
else:
    print("We refuse the project !") 
# it is reverse for the project investment, 
# because we lend our money here, 
# so we want the coupon rate of return to be higher than the required rate of return (ytm).
```
